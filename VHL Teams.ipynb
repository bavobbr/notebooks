```python
import pandas as pd

# Load spreadsheet
xl = pd.ExcelFile('Outdoor Rankings 2022-2023.xlsx')

# Load a sheet into a DataFrame by name
df1 = xl.parse('VHL - KBHB')

df1.columns = ['League', 'Team', 'GS', "WN", "GL", "VL", "PT", "VR", "TG", "PTM"]


# Now you can read values of the DataFrame, for example the cell at row 0, column 0
value = df1["League"]
print(value)
```

```python
col = df1.iloc[:,0].tolist()
print(col)
```

```python
filtered_list = [i for i in col if isinstance(i,str) and not i.isdigit()]
print(filtered_list)
```

```python
for i in filtered_list:
   print(i)
```

```python
leagues_types = list(map(lambda n: n[0:-2], filtered_list))
print(leagues_types)
```

```python
uni_leagues_types = set(leagues_types)
print(uni_leagues_types)
```

```python
leagues_types_uniq = list(dict.fromkeys(leagues_types))
print(leagues_types_uniq)
```

```python
from collections import Counter

count_dict = dict(Counter(leagues_types))
print(count_dict)
```

```python
count_dict_men = {k: v for k, v in count_dict.items() if any(s in k for s in ["Men", "Boys", "Gentleman", "45+"]) }
print(count_dict_men)

count_dict_women = {k: v for k, v in count_dict.items() if any(s in k for s in ["Women", "Girls", "Dames", "Ladies"]) }
print(count_dict_women)
```

```python
for k, v in count_dict_men.items():
    print(f'{k} -> {v}')
```
